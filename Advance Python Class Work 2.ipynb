**3. Write a recursive function to find factorial of a number.**

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
factorial(5)
```

**6. Write a function to check if a number is prime.**

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
is_prime(19)
```

**7. Write a function to find the sum of digits of a number.**

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
sum_of_digits(1234)
```

**8. Write a function to generate first n Fibonacci numbers.**

```python
def generate_fibonacci(n):
    fib = [0, 1]
    if n <= 0: return []
    if n == 1: return [0]

    while len(fib) < n:
        fib.append(fib[-1] + fib[-2])
    return fib
generate_fibonacci(10)
```

**10. Write a function to find length of a string without using len().**

```python
def manual_length(s):
    count = 0
    for _ in s:
        count += 1
    return count
manual_length("Omkar")
```

# **# ROOM ALLOTMENT SYSTEM**

```python
class Hostel:
    def __init__(self, total_rooms):
        self.rooms = {}
        for i in range(1, total_rooms + 1):
            room_id = str(100 + i)
            self.rooms[room_id] = None

    def display(self):
        print("\n--- Current Status ---")
        for room, occupant in self.rooms.items():
            if occupant is None:
                print(f"Room {room}: [Available]")
            else:
                print(f"Room {room}: [Occupied by {occupant}]")

    def allot(self, name):
        for room in self.rooms:
            if self.rooms[room] is None:
                self.rooms[room] = name
                print(f"Room {room} assigned to {name}.")
                return
        print("Sorry, no rooms are left.")

    def vacate(self, room_no):
        if room_no in self.rooms:
            if self.rooms[room_no] is not None:
                print(f"Room {room_no} is now empty.")
                self.rooms[room_no] = None
            else:
                print(f"Room {room_no} was already empty.")
        else:
            print("That room number does not exist.")

my_hostel = Hostel(5)

while True:
    print("\nMenu: 1.View | 2.Allot | 3.Vacate | 4.Exit")
    choice = input("Select an option: ")

    if choice == "1":
        my_hostel.display()
    elif choice == "2":
        student_name = input("Enter Name: ")
        my_hostel.allot(student_name)
    elif choice == "3":
        room_input = input("Enter Room Number: ")
        my_hostel.vacate(room_input)
    elif choice == "4":
        print("Shutting down...")
        break
    else:
        print("Invalid choice.")
```

# **# TRANSACTION SYSTEM**

```python
class Account:
    def __init__(self, pin, balance):
        self.pin = pin
        self.balance = balance

    def deposit(self, amt):
        if amt > 0:
            self.balance += amt
            print("Balance:", self.balance)

    def withdraw(self, amt):
        if 0 < amt <= self.balance:
            self.balance -= amt
            print("Balance:", self.balance)
        else:
            print("Insufficient balance")

    def check(self):
        print("Balance:", self.balance)


accounts = {
    "1001": Account("1234", 5000),
    "1002": Account("5678", 3000),
    "1003": Account("0000", 7000)
}

acc_no = input("Account No: ")

if acc_no in accounts and input("PIN: ") == accounts[acc_no].pin:
    user = accounts[acc_no]

    while True:
        print("\n1.Check 2.Deposit 3.Withdraw 4.Exit")
        ch = input("Choice: ")

        if ch == "1":
            user.check()
        elif ch == "2":
            user.deposit(float(input("Amount: ")))
        elif ch == "3":
            user.withdraw(float(input("Amount: ")))
        elif ch == "4":
            print("Exiting program.  THANK YOU!")
            break
        else:
            print("Invalid choice")
else:
    print("Invalid Account or PIN")
```

# **#1. Parking Lot Management**
"Track vehicle entry/exit, available spots, and compute parking fees"

```python
class MyParkingLot:
    def __init__(self, max_spots):
        self.max_spots = max_spots
        self.parked_cars = {}

    def start_program(self):
        while True:
            cars_inside = len(self.parked_cars)
            print("\n--- PARKING LOT MENU ---")
            print("Total Spots: " + str(self.max_spots))
            print("Cars Parked: " + str(cars_inside))
            print("1. Car Entry")
            print("2. Car Exit (Pay Bill)")
            print("3. Show all parked cars")
            print("4. Close Program")

            choice = input("Select an option: ")

            if choice == "1":
                if len(self.parked_cars) < self.max_spots:
                    plate = input("Enter the car number plate: ")
                    hour = int(input("Enter entry hour (use 1 to 24): "))

                    self.parked_cars[plate] = hour
                    print("Car " + plate + " has parked successfully.")
                else:
                    print("Sorry, the parking lot is full!")

            elif choice == "2":
                plate_to_find = input("Enter the plate of the car leaving: ")

                if plate_to_find in self.parked_cars:
                    entry_hour = self.parked_cars[plate_to_find]
                    exit_hour = int(input("Enter current hour (use 1 to 24): "))

                    hours_stayed = exit_hour - entry_hour

                    if hours_stayed <= 0:
                        hours_stayed = 1

                    total_cost = hours_stayed * 10

                    self.parked_cars.pop(plate_to_find)

                    print("Car left. Stayed for " + str(hours_stayed) + " hour(s).")
                    print("Please pay: ₹" + str(total_cost))
                else:
                    print("That car is not in our parking lot.")

            elif choice == "3":
                if len(self.parked_cars) == 0:
                    print("The lot is empty.")
                else:
                    print("Cars currently inside:")
                    for p in self.parked_cars:
                        print("- " + p)

            elif choice == "4":
                print("Exiting program...")
                break

            else:
                print("Invalid button. Try again.")

my_lot = MyParkingLot(5)
my_lot.start_program()
```

# **#2. To-Do manager with Categories**
"Add task, set deadlines, assign priority and filter by project or tag"

```python
class MyTasks:
    def __init__(self):
        self.all_tasks = []

    def start_manager(self):
        while True:
            print("\n--- MY TO-DO LIST ---")
            print("1. Add a new task")
            print("2. Show all tasks")
            print("3. Search tasks by category (Tag)")
            print("4. Close program")

            choice = input("What do you want to do? ")

            if choice == "1":
                name = input("What is the task? ")
                date = input("When is it due? (e.g. Monday): ")
                level = input("How important is it? (High/Low): ")
                category = input("Enter a tag (Work/Home/School): ")

                new_item = {
                    "title": name,
                    "deadline": date,
                    "priority": level,
                    "tag": category
                }

                self.all_tasks.append(new_item)
                print("Task added to your list!")

            elif choice == "2":

                if len(self.all_tasks) == 0:
                    print("Your list is empty right now.")
                else:
                    print("\n--- ALL TASKS ---")
                    for t in self.all_tasks:
                        print("Task: " + t["title"])
                        print("Due: " + t["deadline"])
                        print("Tag: " + t["tag"])
                        print("-----------------")

            elif choice == "3":
                search_tag = input("Enter the tag you want to find: ")
                found_something = False

                print("\n--- MATCHING TASKS ---")
                for t in self.all_tasks:
                    if t["tag"].lower() == search_tag.lower():
                        print("- " + t["title"] + " (Due: " + t["deadline"] + ")")
                        found_something = True

                if found_something == False:
                    print("No tasks found with that tag.")

            elif choice == "4":
                print("Goodbye!")
                break
            else:
                print("Please pick 1, 2, 3, or 4.")

my_list = MyTasks()
my_list.start_manager()
```

# **#3. Billing System for Retail Store**
"Scane products, apply discount, generate bills and record transaction"

```python
class BillingSystem:
    def __init__(self):
        self.items = {
            "101": ("Apple", 5),
            "102": ("Milk", 20),
            "103": ("Bread", 10)
        }
        self.shopping_cart = []

    def store_open(self):
        while True:
            print("\n--- THE STORE MENU ---")
            print("0. Available items")
            print("1. Add item to cart")
            print("2. Show total and pay")
            print("3. Leave store")

            choice = input("What do you want to do? (0, 1, 2, or 3): ")
            if choice == "0":
                print("\n--- AVAILABLE ITEMS ---")
                for code, item in self.items.items():
                    print(code + ": " + item[0] + " - ₹" + str(item[1]))

            elif choice == "1":
                code = input("Enter the item code: ")

                if code in self.items:
                    product = self.items[code]
                    self.shopping_cart.append(product)
                    print("Success: " + product[0] + " added to your cart.")
                else:
                    print("Error: That item code does not exist.")

            elif choice == "2":
                if len(self.shopping_cart) == 0:
                    print("Your cart is empty. Go buy something first!")
                else:
                    print("\n--- YOUR FINAL BILL ---")
                    total_price = 0

                    for item in self.shopping_cart:
                        name = item[0]
                        price = item[1]
                        print(name + ": ₹" + str(price))
                        total_price = total_price + price

                    print("-----------------------")
                    print("TOTAL TO PAY: ₹" + str(total_price))

                    self.shopping_cart = []
                    print("Thank you for shopping!")
            elif choice == "3":
                print("Goodbye!")
                break
            else:
                print("Invalid choice. Please press 0, 1, 2, or 3.")

my_store = BillingSystem()
my_store.store_open()
```

# **#4. Movie Ticket Book tickets**
" Check showtimes, book tickets select seats and print"

```python
class MyCinema:
    def __init__(self):
        self.all_seats = [
            ["O", "O", "O", "O", "O"],
            ["O", "O", "O", "O", "O"],
            ["O", "O", "O", "O", "O"],
            ["O", "O", "O", "O", "O"],
            ["O", "O", "O", "O", "O"]
        ]

    def show_the_map(self):
        print("\n   1 2 3 4 5")
        row_number = 1
        for row in self.all_seats:
            print(str(row_number) + "  " + " ".join(row))
            row_number = row_number + 1

    def start_booking(self):
        while True:
            print("\n--- MOVIE BOOKING ---")
            print("1. View Seats")
            print("2. Book Seat")
            print("3. Exit")

            choice = input("Choice: ")

            if choice == "1":
                self.show_the_map()

            elif choice == "2":
                r = int(input("Row (1-5): ")) - 1
                c = int(input("Col (1-5): ")) - 1

                if r >= 0 and r < 5 and c >= 0 and c < 5:
                    if self.all_seats[r][c] == "O":
                        self.all_seats[r][c] = "X"
                        print("Confirmed.")
                    else:
                        print("Taken.")
                else:
                    print("Invalid seat.")

            elif choice == "3":
                break

cinema = MyCinema()
cinema.start_booking()
```
