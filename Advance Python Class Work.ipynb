```python
# List based transaction system
balance = 0
transactions = []

a = 1
password = int(input("Enter your password: "))
while True:
    if password == 1415:
        while True:
         print("\n.1.Depodit:")
         print("2.withdraw:")
         print("3.check balance:")
         print("4.transaction history:")
         print("5.exit.")

         choice = int(input("Enter your choice: "))
         if choice == 1:
              amt = int(input("Enter the amount to deposit: "))
              balance += amt
              transactions.append("Deposited"+str(amt)+"only")
              print("Amount Deposited")
         elif choice == 2:
              amt = int(input("Enter the amount to withdraw: "))
              if amt <= balance:
                  balance -=amt
                  transactions.append("Withdraw "+str(amt)+" only")
                  print("Amount Withdraw")
              else:
                  print("Insufficient balance")
         elif choice == 3:
              print("Current balance:", balance)
         elif choice == 4:
              print("Transaction history:")
              for transaction in transactions:
                  print(transaction)
         else:
              print("Exit")
    else:
        a_left = 3 - a
        a += 1
        if a_left == 0:
            for i in range(30, 0, -1):
                print(f"Trying again in {i} seconds...", end='\r')
                import time
                time.sleep(1)
            a = 1
            a_left = 3 - a
            print("\nEnter your password again: ")
            password = int(input())
        else:
            print("Attempt left: " + str(a_left))
            password = int(input("Enter your password: "))
```

**2# Online Shopping Cart**

```python
# Initial Setup
stock = {
    "Laptop": 5,
    "Mouse": 10,
    "Keyboard": 3
}

prices = {
    "Laptop": 1000,
    "Mouse": 25,
    "Keyboard": 50
}

cart = {}

while True:
    print("\n--- Items in Stock ---")
    for item, qty in stock.items():
        print(f"{item}: {qty} available (${prices[item]})")

    print("\nOptions: [add] to cart, [buy] now, [view] cart, or [exit]")
    choice = input("What would you like to do? ").lower()

    if choice == "exit":
        print("Goodbye!")
        break

    elif choice == "add":
        item = input("Enter item name to add: ").capitalize()
        if item in stock and stock[item] > 0:
            cart[item] = cart.get(item, 0) + 1
            print(f"{item} added to cart.")
        else:
            print("Sorry, item out of stock or does not exist.")

    elif choice == "buy":
        item = input("Enter item name to buy immediately: ").capitalize()
        if item in stock and stock[item] > 0:
            stock[item] -= 1
            print(f"{item} Purchase successful! ")
        else:
            print("Transaction failed. Item unavailable.")

    elif choice == "view":
        print("\n--- Your Cart ---")
        if not cart:
            print("Empty")
        else:
            for item, qty in cart.items():
                print(f"{item} x{qty}")

    else:
        print("Invalid option, try again.")

```

**3# Library Management**

**# Traffic Light Simulation**

```python
# Initial Setup
# {Book Title: Number of copies available}
library_stock = {
    "Harry Potter": 3,
    "The Hobbit": 2,
    "Demon Slayer": 5,
    "Python Basics": 1
}

# Member's collection (Books currently held)
my_books = []

# Main Loop
while True:
    print("\n--- Library Books Available ---")
    for book, count in library_stock.items():
        status = "Available" if count > 0 else "Out of Stock"
        print(f"📖 {book} [{count} copies] - {status}")

    print("\nOptions: [borrow], [return], [view] my books, or [exit]")
    choice = input("What would you like to do? ").lower()

    if choice == "exit":
        print("Closing Library System. Goodbye!")
        break

    elif choice == "borrow":
        book_name = input("Enter the book title to borrow: ").strip().title()

        if book_name in library_stock:
            if library_stock[book_name] > 0:
                # Logic: Reduce from library, add to member list
                library_stock[book_name] -= 1
                my_books.append(book_name)
                print(f"✅ Success! You have borrowed '{book_name}'.")
            else:
                print(f"❌ Sorry, all copies of '{book_name}' are currently out.")
        else:
            print("❌ That book is not in our catalog.")

    elif choice == "return":
        book_name = input("Enter the book title to return: ").strip().title()

        if book_name in my_books:
            # Logic: Add back to library, remove from member list
            library_stock[book_name] += 1
            my_books.remove(book_name)
            print(f"✅ Thank you! '{book_name}' has been returned.")
        else:
            print("❌ You don't have that book in your possession.")

    elif choice == "view":
        print("\n--- Your Borrowed Books ---")
        if not my_books:
            print("You haven't borrowed any books yet.")
        else:
            for i, book in enumerate(my_books, 1):
                print(f"{i}. {book}")

    else:
        print("⚠️ Invalid option. Please type 'borrow', 'return', 'view', or 'exit'.")
```
