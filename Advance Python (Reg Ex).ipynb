```python
import re

pattern = "Cat"
text = "A Cat is dancing"

match = re.search(pattern, text)
print(match)
print(type(match))
print(match.start())
print(match.end())
print(match.span())
```

```python
### 1. Find the largest element in a list

lst = [3, 7, 2, 9, 5]
largest = lst[0]
for num in lst:
    if num > largest:
        largest = num
print(largest)
```

```python
### 2. Check if a number is even or odd

n = 7
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
### 3. Reverse a string (without slicing)

s = "hello"
rev = ""
for ch in s:
    rev = ch + rev
print(rev)
```

```python
### 4. Count vowels in a string

s = "education"
count = 0
for ch in s.lower():
    if ch in "aeiou":
        count += 1
print(count)
```

```python
### 5. Find the sum of digits of a number

n = 1234
total = 0
while n > 0:
    total += n % 10
    n //= 10
print(total)
```

```python
### 6. Check if a string is a palindrome

s = "madam"
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
### 7. Print Fibonacci series up to N terms

n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
### 8. Swap two variables without a third variable

a, b = 5, 10
a, b = b, a
print(a, b)
```

```python
### 9. Count words in a sentence

sentence = "Python is fun to learn"
words = sentence.split()
print(len(words))
```

```python
### 10. Find minimum and maximum in a list

lst = [4, 1, 9, 7]
print(min(lst), max(lst))
```

```python
### 11. Find the second largest number in a list

lst = [10, 20, 4, 45, 99]
lst.sort()
print(lst[-2])
```

```python
### 12. Remove duplicates from a list (without set)

lst = [1, 2, 2, 3, 4, 4]
unique = []
for num in lst:
    if num not in unique:
        unique.append(num)
print(unique)
```

```python
### 13. Count frequency of elements in a list

lst = [1, 2, 2, 3, 1]
freq = {}
for num in lst:
    freq[num] = freq.get(num, 0) + 1
print(freq)
```

```python
### 14. Merge two sorted lists
a = [1, 3, 5]
b = [2, 4, 6]
merged = sorted(a + b)
print(merged)
```

```python
### 15. Find the intersection of two lists

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
intersection = [x for x in a if x in b]
print(intersection)
```

```python
### 16. Rotate a list by k positions
lst = [1, 2, 3, 4, 5]
k = 2
rotated = lst[k:] + lst[:k]
print(rotated)
```

```python
### 17. Check if two strings are anagrams
s1, s2 = "listen", "silent"
if sorted(s1) == sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```

```python
### 18. Find the first non-repeating character
s = "swiss"
for ch in s:
    if s.count(ch) == 1:
        print(ch)
        break
```

```python
### 19. Flatten a nested list
nested = [[1, 2], [3, 4], [5]]
flat = []
for sub in nested:
    for item in sub:
        flat.append(item)
print(flat)
```

```python
### 20. Find all pairs with a given sum
lst = [1, 2, 3, 4, 5]
target = 6
pairs = []
for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == target:
            pairs.append((lst[i], lst[j]))
print(pairs)
```

```python
### 21. Print a right triangle pattern

n = 5
for i in range(1, n+1):
    print("*" * i)
```

```python
### 22. Print a pyramid pattern

n = 5
for i in range(1, n+1):
    print(" "*(n-i) + "$"*(2*i-1))
```

```python
### 23. Print all prime numbers in a range

start, end = 10, 50
for num in range(start, end+1):
    if num > 1:
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

```python
### 24. Check if a number is Armstrong

n = 153
digits = str(n)
total = sum(int(d)**len(digits) for d in digits)
if total == n:
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python
n = int(input('Enter a number: '))
digi = str(n)
tota = sum(int(i)**len(digi) for i in digi)

if tota == n:
    print('It is an Armstrong number.')
else:
    print('It is not an Armstrong number.')
```
